### 11. How many students got A, B, C, F?

```python
dic={'name':['samad','anas','fasih','rafay','farukh','yasoob'],'marks':[80,80,79,60,20,22],'id':[1,2,3,4,5,6]}
data=pd.DataFrame(dic)
#a=(data[data.marks>40])
count={}
#display(len(a))
for a in data.marks:
    a=int(a)
    if a >= 80 and  a <= 90 :
        s='A'
        count[s]=count.get(s,0)+1
    elif a >= 70 and  a <= 79 :
        s='B'
        count[s]=count.get(s,0)+1
    elif a >= 60 and  a <= 69 :
        s='C'
        count[s]=count.get(s,0)+1
    else:
        s='F'
        count[s]=count.get(s,0)+1
            
print(count)
```

### 12. What are the mean scores for students who got A, B, C, F?

```python
dic={'name':['samad','anas','fasih','rafay','princexx','alina'],'marks':[80,80,79,60,20,22],'id':[1,2,3,4,5,6]}
data=pd.DataFrame(dic)
a=(data[data.marks>40])
count={}
#display(len(a))
for a in data.marks:
    a=int(a)
    if a >= 80 and  a <= 90 :
        s='A'
        count[s]=count.get(s,0)+1
    elif a >= 70 and  a <= 79 :
        s='B'
        count[s]=count.get(s,0)+1
    elif a >= 60 and  a <= 69 :
        s='C'
        count[s]=count.get(s,0)+1
    else:
        s='F'
        count[s]=count.get(s,0)+1           
print(count)
sum=0
c=0
for i,h in count.items():
    h=int(h)
    sum=sum+h
    c+=1
print("Mean",sum/c)
```
